```python
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import SnowballStemmer
```

```python
stop_words = stopwords.words('french')
stop_words += stopwords.words('english')
stop_words
```

```python
def tok(sentence):
    return nltk.word_tokenize(sentence.lower())


def no_stop(tokens):
    return [token for token in tokens if (token not in stop_words) and (token.isalpha())]

def stem(tokens, language='french'):
    stemmizer = SnowballStemmer(language=language)
    return [stemmizer.stem(token) for token in tokens]

def preprocess(sentence, stemm=True, stop=True):
    if stemm:
        return ' '.join(stem(no_stop(tok(sentence)), language='english'))
    else:
        return ' '.join(no_stop(tok(sentence)))

df2["processed"] = df2['tags'].apply(lambda x : preprocess(x, stemm=False))
df2
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid_vectorizer = TfidfVectorizer()

X = df2['processed']

tfid_vectorizer.fit(X)

X_tfid = tfid_vectorizer.transform(X)
```

```python
tfid_vectorizer.get_feature_names_out()
```

```python
pd.DataFrame(X_tfid.toarray(), columns = tfid_vectorizer.get_feature_names_out())
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
modelNN_t = NearestNeighbors(n_neighbors=3, metric='cosine')
modelNN_t.fit(X_tfid)
```

```python
modelNN_t.kneighbors(X_tfid)[1]
```

```python
def recommend(festival):
    # Transform the given movie's tag into TF-IDF vector
    festival_tag_tfidf = tfid_vectorizer.transform([festival])
    
    # Find indices of k-nearest neighbors (excluding the movie itself)
    distances, indices = modelNN_t.kneighbors(festival_tag_tfidf, n_neighbors=6)
    indices = indices.flatten()[1:]  # Exclude the movie itself
    
    # Print recommended movies
    print("Recommended Movies:")
    for i, idx in enumerate(indices):
        nom_festival = df2.iloc[idx].Nom_festival
        print(f"{i+1}. {nom_festival}")
```

```python
recommend('Avoine Zone Groove')
```
